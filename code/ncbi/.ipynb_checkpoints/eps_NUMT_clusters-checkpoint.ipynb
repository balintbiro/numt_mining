```python
#import dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
features=pd.read_csv('../../data/order_umap.csv',index_col=0)
```

```python
fig,axs=plt.subplots(figsize=(8.5,6.5))
kwargs={'edgecolor':'face'}
scplot=sns.scatterplot(x='x',y='y',hue='order',data=features,s=10,palette='tab20',alpha=1,**kwargs)
axs.set(xticklabels=[],yticklabels=[])
#scplot.legend(title='Sequence',fontsize=15,title_fontsize=15)#this is for the numt vs random umap
axs.set_xlabel('UMAP1',fontsize=20)
axs.set_ylabel('UMAP2',fontsize=20)
axs.legend(
    bbox_to_anchor=(1,1.0125),title='Orders',prop={'size': 12},
    markerscale=2,fontsize=20
)
plt.tight_layout()
plt.savefig('../../results/NUMT_clusters.eps',dpi=400)
```
