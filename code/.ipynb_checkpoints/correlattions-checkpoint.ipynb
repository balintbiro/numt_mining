```python
#importing required modules
import os
import numpy as np
import scipy.stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#get nuclear genome informations
gdnas=pd.read_csv('../data/eukaryotes.txt',sep='\t',index_col=0)
gdnas.head(n=2)
```

```python
gdnas.loc[gdnas['BioProject ID']==33554]
```

```python
#get mitochondrial genome informations
mtdnas=pd.read_csv('../data/organelles.csv',index_col=0)
mtdnas.head(n=2)
```

```python
#get phylogenetic informations
phylo_data=pd.read_csv('../data/nodes.dmp',sep='\t',header=None)
phylo_data=phylo_data.rename(columns={
    0:'tax_id',
    4:'rank'
                          })
phylo_data.head(n=2)
```

```python
#add phylogenetic rank to gdnas based on tax_id
def add_phylo_rank(tax_id):
    try:
        rank=phylo_data.loc[phylo_data['tax_id']==tax_id]['rank'].tolist()[0]
        return rank
    except:
        return np.nan

ranks=gdnas['TaxID'].apply(add_phylo_rank)
gdnas['rank']=ranks
gdnas.head(n=2)
```

```python
#function for adding mtdna size to gdnas
def get_mt_size(gdna_row):
    mtdna=mtdnas.loc[mtdnas['BioProject']==gdna_row['BioProject Accession']]
    try:
        return mtdna['Size(Mb)'].tolist()[0]
    except:
        return np.nan
    
mt_dna_sizes=gdnas.apply(get_mt_size, axis=1)
```

```python
#add mtdna size to gdna df
gdnas['mt_dna_size(Mb)']=mt_dna_sizes
gdnas.head(n=2)
```

```python
#drop rows where no data for genome sizes
gdnas=gdnas.dropna(subset=['Size (Mb)','mt_dna_size(Mb)'])
gdnas.head(n=2)
```

```python
#read in numts
numts=pd.read_csv('../results/tsne_input.csv')
```

```python
#get numt counts for every organism
numt_counts=pd.Series(numts['organism_name'].unique()).apply(
    lambda organism_name:len(numts.loc[numts['organism_name']==organism_name])
)
```

```python
#get numt lengths for every organism
numt_lengths=pd.Series(numts['organism_name'].unique()).apply(
    lambda organism_name:sum(numts.loc[numts['organism_name']==organism_name]['genomic_length'])
)
```

```python
#get genome sizes for every organisms
genome_sizes=pd.Series(numts['organism_name'].unique()).apply(
    lambda organism_name:numts.loc[numts['organism_name']==organism_name]['full_g_size'].tolist()[0]
)
```

```python
scipy.stats.spearmanr(genome_sizes,numt_counts)
```

```python
#calculate correlation WITHOUT genome quality control
g_gc=numts.drop_duplicates(subset=['organism_name'])['g_gc']
plt.style.use('fivethirtyeight')
fig,axs=plt.subplots(1,1,figsize=(5,5))
axs.scatter(g_gc,numt_lengths,edgecolors='black',s=200,alpha=.25)
m, b = np.polyfit(g_gc,np.log(numt_lengths),1)
axs.plot(g_gc, np.exp(m*g_gc + b),'r',linewidth=1)
axs.set_yscale('log')
axs.set_xlabel('gDNS GC arány',fontsize=30)
axs.set_ylabel('Kumulált numt hossz\n(bp)',fontsize=30)
plt.tight_layout()
plt.savefig('../results/gc_numt_length_corr.png',dpi=450)
```

```python
#calculate correlation WITHOUT genome quality control
fig,axs=plt.subplots(1,1,figsize=(5,5))
axs.scatter(genome_sizes,numt_counts,edgecolors='black',s=200,alpha=.25)
m, b = np.polyfit(genome_sizes,np.log(numt_counts),1)
axs.plot(genome_sizes, np.exp(m*genome_sizes + b),'r',linewidth=1)
axs.set_xlabel('Genom méret (Mb)',fontsize=30)
axs.set_ylabel('Numt szám (db)',fontsize=30)
axs.set_yscale('log')
plt.tight_layout()
plt.savefig('../results/gsize_numt_count_corr.png',dpi=450)
```

```python
#calculate GC ratio for numt corresponding mt sequence
numt_corr_gc=numts['mitochondrial_sequence'].apply(lambda seq:(seq.upper().count('G')+seq.upper().count('C'))/len(seq.replace('N','')))
numts['numt_corr_ggc']=numt_corr_gc
```

```python
matrix_input=numts[['organism_name','score','eg2_value','e_value','genomic_length','numt_gc','full_mt_size','full_g_size','numt_corr_ggc']]
```

```python
g_mt_sizes=numts[['organism_name','full_mt_size','full_g_size','g_gc','mt_gc']].drop_duplicates()
g_mt_sizes.index=g_mt_sizes['organism_name']
g_mt_sizes
```

```python
def normalize(organism_name,column_to_normalize):
    global rel_numt_sizes
    global rel_numt_ggc
    global rel_numt_mtgc
    subdf=matrix_input.loc[matrix_input['organism_name']==organism_name]
    if column_to_normalize=='genomic_length':
        rel_numt_sizes+= (subdf['genomic_length']/(g_mt_sizes.loc[organism_name]['full_mt_size']*1000000)).tolist()
    else:
        rel_numt_ggc+= ((subdf['numt_gc']*100)/g_mt_sizes.loc[organism_name]['g_gc']).tolist()
        rel_numt_mtgc+= ((subdf['numt_corr_ggc']*100)/g_mt_sizes.loc[organism_name]['mt_gc']).tolist()
```

```python
rel_numt_sizes=[]
pd.Series(numts['organism_name'].unique()).apply(normalize,args=('genomic_length',))
rel_numt_sizes[:5]
```

```python
rel_numt_mtgc=[]
rel_numt_ggc=[]
pd.Series(numts['organism_name'].unique()).apply(normalize,args=('g_gc',))
rel_numt_ggc[:5]
```

```python
numts['rel_numt_size']=rel_numt_sizes
numts['rel_numt_ggc']=rel_numt_ggc
numts['rel_numt_mtgc']=rel_numt_mtgc
```

```python
numts[numts['rel_numt_size']>1]['organism_name']
```

```python
kwargs ={"color":"brown",'alpha':.4}
color='black'
fig,ax1=plt.subplots()
ax1.hist(numts[numts['rel_numt_size']<.05]['rel_numt_size'],bins=200,**kwargs,zorder=3)
l, b, h, w = .7, .45, .45, .3
ax2 = fig.add_axes([l, b, w, h])
ax=ax2.boxplot(numts['genomic_length'],showfliers=False,
           boxprops={'linewidth':3,'color':color,'alpha':.4},
           whiskerprops={'linewidth':3,'color':color,'alpha':.4},
           capprops={'linewidth':3,'color':color,'alpha':.4},
           patch_artist=True)
for patch in ax['boxes']:
    patch.set_facecolor('brown')
right_side = ax2.spines["right"]
right_side.set_visible(False)
top_side = ax2.spines["top"]
top_side.set_visible(False)
ax2.set_xticklabels([])
plt.tight_layout()
ax1.set_xlabel('Relatív numt\nméret',fontsize=20)
ax1.set_ylabel('Gyakoriság',fontsize=20)
ax2.set_ylabel('Numt méret\n(bp)',fontsize=20)

plt.tight_layout()
plt.savefig('../results/ncbi_rel_numt_size.png',dpi=450)
```

```python
kwargs ={"color":"blue",'alpha':.4,'label':'gDNS'}
ax=numts['rel_numt_ggc'].hist(bins=200,**kwargs)
ax.set_xlabel('Relatív GC arány',fontsize=20)
kwargs ={"color":"green",'alpha':.4,'label':'mtDNS'}
numts['rel_numt_mtgc'].hist(bins=200,**kwargs)
ax.set_ylabel('Gyakoriság',fontsize=20)
plt.legend(fontsize=20)
plt.tight_layout()
plt.savefig('../results/ncbi_rel_gcs.png',dpi=450)
```
