```python
#import dependencies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
#read in dataframe
#import numts
numts=pd.read_csv('../data/ncbi_numts_modk2_added.csv',index_col=0)
numts['score']=numts.index
numts.index=np.arange(0,len(numts))
```

```python
#create datasets
X=numts[[
    'score','eg2_value','e_value','genomic_start','genomic_length','mitochondrial_length','GC'
        ]]#,'modk2'
y=numts['organism_label']
```

```python
X_normalized=StandardScaler().fit(X).transform(X)
tsne = TSNE(random_state = 0,perplexity=60,n_jobs=-1)
X_tsne = tsne.fit_transform(X_normalized)
```

```python
X['organism_name']=numts['organism_name']
```

```python
X['x']=X_tsne[:,0]
X['y']=X_tsne[:,1]
X['label']=y
plt.style.use('fivethirtyeight')
fig,axs=plt.subplots(1,1,figsize=(10,10))
sns.scatterplot(
    x='x',
    y='y',
    hue='label',
    data=X,
    palette='Paired',
    alpha=.7,
    ax=axs
)
axs.axis('off')
plt.legend().remove()
plt.tight_layout()
plt.xlim(45,62.5)
plt.ylim(0,17.5)
plt.savefig('../results/ncbi_pp47_to_annotate_zoomed.png',dpi=450)
```

```python
def get_organism_name(input_df,xmin,xmax,ymin,ymax):
    fil1=input_df['x']<xmax
    X1=input_df[fil1]
    fil2=X1['x']>xmin
    X2=X1[fil2]
    fil3=X2['y']>ymin
    X3=X2[fil3]
    fil4=X['y']<ymax
    X4=X3[fil4]
    return X4['organism_name'].value_counts()
get_organism_name(X,52,54,4,6)
```
