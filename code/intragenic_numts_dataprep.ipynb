```python
#import required modules
import os
import numpy as np
import pandas as pd
```

```python
#read in the csv files
parent_dir=os.path.join('/Volumes/motilin/balint/numt/mice_strains_numt/gene_ids/')
dfs=pd.Series(os.listdir('/Volumes/motilin/balint/numt/mice_strains_numt/gene_ids/')).apply(
lambda filename: pd.read_csv(parent_dir+filename) if filename.count('.')==1 else print(filename))
dfs.index=pd.Series(os.listdir('/Volumes/motilin/balint/numt/mice_strains_numt/gene_ids/')).apply(
    lambda filename:filename[:-4])
dfs=dfs.drop(['._mm','._mmspretus'])
```

```python
dfs
```

```python
#get the genes without repetition for each strain
def get_unique_genes(df):
    global unique_genes
    unique_genes+=list(df['values'].unique())
    
unique_genes=[]
dfs.apply(get_unique_genes)
unique_genes=pd.Series(unique_genes)
gene_frequencies=unique_genes.value_counts()
#gene count 16 means that the given intragenic numt is present in every strain
gene_frequencies[gene_frequencies>15]
```

```python
#add strain name to strain specific dataframes
def add_strain_name(df):
    global df_indexer
    strain=dfs.index.values[df_indexer]
    df_indexer+=1
    df['strain']=len(df)*[strain]
    
    
df_indexer=0
dfs.apply(add_strain_name)
```

```python
#merge the dataframes
merge_input=[]
dfs.apply(lambda df: merge_input.append(df))

merged_df=pd.concat(merge_input)
merged_df
```

```python
#check the count for every gene in every strains
def gene_checker(gene):
    subdf=merged_df.loc[merged_df['values']==gene]
    counts=subdf['strain'].value_counts()
    return(counts)
gene_counts=pd.Series(merged_df['values'].unique()).apply(gene_checker)
gene_counts.index=merged_df['values'].unique()
gene_counts
```

```python
#all genes through all strains
mm_nas=gene_counts.apply(lambda row:row['mm'],axis=1)
mm_nas
```

```python
#get the gene names which are present in mm
mm_genes=mm_nas.dropna()
mm_genes
```

```python
#get the genes that are present in mm and not mitochondrial genes
gene_mask=pd.Series(list(set(mm_nas.index.values)-set(mm_genes.index.values))).apply(lambda gene: 'mt' not in gene)
#genes of interest are the genes that are present in other strains but not in mm and not mitochondrial genes
gois=pd.Series(list(set(mm_nas.index.values)-set(mm_genes.index.values)))[gene_mask]
gois
```

```python
gois_count=gois.apply(lambda goi: gene_counts.loc[goi])
gois_count.index=gois
gois_count
```

```python
#https://www.nature.com/articles/s41588-018-0223-8#Sec1
wild_derived_mice=pd.Series(['mm','mmcasteij','mmpwkphj','mmspretus','mmwsbeij'])
wild_derived_mice.index=['mm','mmcasteij','mmpwkphj','mmspretus','mmwsbeij']
wild_derived_mice.apply(lambda strain: gois_count[strain]).transpose()
```
